# LDA(Latent Dirichlet Allocation)
- LDA(잠재 디리클레 할당)의 목표: 문서의 집합에서 토픽을 찾아내는 것이 목표이며, 이것을 토픽 모델링이라고 함.

- 문서들은 주제들의 혼합으로 구성되어 있고, 주제들은 확률 분포에 기반하여 단어들을 생성한다고 가정한다.

- 데이터가 주어지면, LDA는 문서가 생성되던 과정을 역추정함.

- LDA(잠재 디리클레 할당)의 장점: 주제 벡터를 구하는 한 방법으로, LSA보다 사람이 해석하기 좋은 형태로 주제 벡터를 만들어낸다.
  - LDA(Latent Dirichlet Allocation)이 LSA보다 직관적이다.
  - LSA는 선형적 관계로만 나타나기 때문에 해석이 힘들다.

- LDA는 문서들이 주제들의 혼합으로 이루어져 있다고 가정하며, 또한 주제들은 확률 분포에 기반하여 각 주제와 연관된 단어를 생성한다고 가정한다.
  - k개의 주제들 -> 가중치 summation = 1.0
  - 주어진 문서를 바탕으로 주제와 단어의 확률 분포를 역추적해나가는 것이 LDA의 핵심이다.
  - LDA는 선형적이지 않고 비선형적으로 변환됨. -> 선형적으로 설명 X
  - LDA는 거꾸로 분포를 추적해 나가야함.

-  LDA(Latent Dirichlet Allocation)에서는 아래의 과정을 통해 문서가 생성된다고 가정함.
  - 1. 문서에 사용할 토큰의 개수 n을 정함.
  - 2. 문서에 사용할 주제의 개수 및 혼합 비율을 정함.
  - 3. 문서에 사용할 각 토큰을 아래와 같이 선정함.
    - 3-1. 주제를 확률 분포에 따라 선택
    - 3-2. 선택된 주제에서의 토큰 확률 분포에 따라 최종 토큰 선택

- LDA 예시
  - 문서1: 저는 사과랑 바나나를 먹어요.
  - 문서2: 우리는 귀여운 강아지가 좋아요.
  - 문서3: 저의 깜찍하고 귀여운 강아지가 바나나를 먹어요.

  - 1. 토픽 개수(n) 가정: n = 2
    - n 하이퍼 파라미터의 설정에 따라 결과가 좋을수도 나쁠수도 있다.

  - 2. 각 문서의 토픽 분포 구하기 -> 모델이 혼합 비율을 반복적인 샘플링과 추정 과정을 통해 학습하여 결정함.
    - 2-1.초기화 단계
      - LDA는 각 문서에 대해 랜덤하게 토픽을 할당하거나, 일부 초기 추정치를 사용하여 토픽 할당을 시작한다.
    - 2-2. 반복적 갱신(Gibbs 샘플링)
      - 혼합 비율을 점진적으로 업데이트한다. 이 과정은 수렴할 때까지 반복된다.
      - 샘플링은 각 단어가 특정 토픽에 속할 확률을 반복적으로 샘플링하면서, 토픽 분포와 단어 분포를 업데이트한다.
      - 각 단어의 토픽 할당을 업데이트할 때, LDA는 1. 문서에서 해당 토픽이 나타날 확률,2. 해당 토픽 내에서 해당 단어가 나타날 확률을 고려한다.
    - 2-3. 토픽 분포와 단어 분포 계산
      - 샘플링이 반복되면서 각 문서의 토픽 할당과 각 토픽의 단어 할당이 갱신된다.
      - 이 데이터를 바탕으로 문서의 토픽 분포는 문서 내에서 각 토픽이 할당된 단어의 비율로 계산된다.
      - ex) 문서에 100개의 단어가 있고, 그 중 30개가 토픽 1에 속하면, 토픽 1의 비율은 30% 이다.
    - 2-4. 수렴
      - 위의 과정이 반복되면서, 각 문서의 토픽 분포와 각 토픽의 단어 분포는 점차 안정된 상태(수렴)에 도달하게 됨.
      - 이 수렴된 상태에서는 LDA가 학습한 각 문서의 토픽 혼합 비율과 각 토픽이 단어 혼합 비율을 확률 값으로 표현함.

    - 문서1 토픽: 과일 100%
    - 문서2 토픽: 동물 100%
    - 문서3 토픽: 동물 60%, 과일 40%

  - 3. 각 토픽의 단어 분포
    - 3-1. 각 문서의 토큰화 진행
    - 3-2. 토큰 = [사과, 바나나, 먹어요, 귀여운, 강아지, 깜찍하고, 좋아요]
    - 3-3. 각 문서당 BoW 형식으로 나타냄.
      - Token= [사과, 바나나, 먹어요, 귀여운, 강아지, 깜찍하고, 좋아요]
      - Doc1 = [   1,      1,      1,      0,      0,        0,      0]
      - Doc2 = [   0,      0,      0,      1,      1,        0,      1]
      - Doc3 = [   0,      1,      1,      1,      1,        1,      0]
      - summation = [1, 2, 2, 2, 2, 1, 1]
      - 토픽 1 안에는 [사과, 바나나, 먹어요]만 생각
      - 토픽 2 안에는 [귀여운, 강아지, 깜찍하고, 좋아요]만 생각
      - len([사과, 바나나, 먹어요]) = 5
      - len([귀여운, 강아지, 깜찍하고, 좋아요]) = 6
      - <각 토픽의 단어 분포>
        - 토픽1: [1/5, 2/5, 2/5, 0, 0, 0, 0]
        - 토픽2: [0, 0, 0, 2/6, 2/6, 1/6, 1/6]

- LDA 과정 요약
  - 1. 토픽 개수 k 설정
  - 2. 문서에 있는 모든 단어를 k개의 토픽 중 하나에 랜덤으로 할당
  - 3. 각 문서의 개별 단어에 대해 나머지 단어들은 모두 제대로 할당되어 있다고 가정하고, 토픽을 재할당하며, 수렴시까지 이를 반복함.     

```python
# LDA(Latent Dirichlet Allocation) 실습

# 데이터 셋 로드 및 BoW 생성
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize.casual import casual_tokenize
import numpy as np
np.random.seed(42)
```

```python
sms = pd.read_csv('sms-spam.csv', index_col=0)
index = [f'sms{i}{"!"*j}' for (i, j) in zip(range(len(sms)), sms.spam)]
sms.index = index
counter = CountVectorizer(tokenizer=casual_tokenize)
```

```python
bow_docs = pd.DataFrame(counter.fit_transform(raw_documents=sms.text).toarray(), index=index)
```

```python
column_nums, terms = zip(*sorted(zip(counter.vocabulary_.values(), counter.vocabulary_.keys())))
```

```python
bow_docs.columns = terms
```

```python
print(sms.loc['sms0'].text)
```

```python
print(bow_docs.loc['sms0'][bow_docs.loc['sms0'] > 0].head())
```

```python
bow_docs.head()
```

```python
# LDA 수행 및 단어 - 주제 벡터 확인
from sklearn.decomposition import LatentDirichletAllocation as LDA

lda = LDA(n_components=16, learning_method='batch')
lda = lda.fit(bow_docs)
```

```python
print(lda.components_.shape)
```

```python
columns = [f'topic{i}' for i in range(lda.n_components)]
components = pd.DataFrame(lda.components_.T, index=terms, columns = columns)
```

```python
print(components.round(2).head(3)) # 정규화된 BoW 형식
```

```python
# Topic3에 대한 해석 시도 - 강조, 명령 등에 사용되는 단어들이 순위가 높음.
components.topic3.sort_values(ascending=False)[:10] # 내림차순 정렬
```

```python
# LDA 기반의 문서 - 주제 벡터 생성
lda_topic_vectors = lda.transform(bow_docs)
lda_topic_vectors = pd.DataFrame(lda_topic_vectors, index=index, columns=columns)
```

```python
lda_topic_vectors.round(2).head() # 각 문서의 topic 혼합 비율
```

```python
# 생성된 주제 벡터를 이용하여, 선형 판별 분석 수행
# 스팸 분류와 같은 불균형 데이터 셋에서는 정밀도와 재현율이 중요하며 이 모두를 고려한 F1 score가 모델 성능 지표로 널리 활용됨.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

X_train, X_test, y_train, y_test = train_test_split(lda_topic_vectors, sms.spam, test_size=0.5, random_state=42) # train, test가 1:1로 나눈다.
linear_discriminant_analysis = LinearDiscriminantAnalysis(n_components = 1)
linear_discriminant_analysis = linear_discriminant_analysis.fit(X_train, y_train)
train_y_pred = linear_discriminant_analysis.predict(X_train)
test_y_pred = linear_discriminant_analysis.predict(X_test)
```

```python
# LDA를 활용한 F1-score
print("Train F1-score:", f1_score(y_train, train_y_pred))
print("Test F1-score:", f1_score(y_test, test_y_pred))
# F1-score를 통해 PCA가 LDA보다 성능이 더 좋다는 것을 알 수 있다.
```

# LDA(Latent Dirichlet Allocation) vs LSA

- LSA: SVD, PCA 방식 모두 선형 분석을 수행함.
- LDA: 비선형적인 방식으로 주제 벡터를 선출함.
- 주제 벡터 유사도 등 연산에 활용함에 있어서는 LSA가 전반적으로 LDA보다 좋은 성능을 보인다.

- LDA 특징
  - 1. 사람이 해석하기에 용이한 형태로 주제 벡터를 산출하는 경향이 있음
  - 2. LDA의 문서-주제 벡터를 살펴보면, LDA에서의 결과가 좀 더 sparse(희소함)한 것을 알 수 있다.
  - 3. LSA에 비해 LDA는 0의 값을 갖는 주제가 많아 각 문서를 좀 더 적은 개수의 주제들로 표현한다.
  - 4. LDA는 문서 집합에 대해서 토픽들을 추출하고 해석하는 토픽 모델링 분야를 개척한 기술이자 핵심 기술로써 사용되어 왔다.
